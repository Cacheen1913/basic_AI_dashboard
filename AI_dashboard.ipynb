```python
import streamlit as s
import google.generativeai as genai
import pyttsx3
import time

key = "here enter your Gemini API key"
genai.configure(api_key = key)
model = genai.GenerativeModel("gemini-2.5-flash")

s.title("Problem Solver")
s.header("Welcome Sir")
s.subheader("How can i assist you today?")
s.write("Anything you can ask me, I am here to help you")
ip = s.text_input("Here you can type!")
response = ""

if s.button("EXIT"):
    s.success("Bye Bye! Will meet soon...")
if ip:
    # if ip.lower() == "exit":
    #     s.write("Bye Bye, Will meet soon!")
    #     s.stop()
    #else:
    response = model.generate_content(ip.lower())
    s.write("Here You Go : ", response.text)
    speaker = pyttsx3.init()
    speaker.say(response.text)
    speaker.runAndWait()
    del speaker
        
```
